```python
import pandas as pd

# Define a function to convert Stata output text file to a pandas DataFrame
def stataConverter(file_path):
    # Open the specified text file and read its lines
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    # Extract the numerical part from the file name
    num = file_path.strip(".txt").strip("column")
    
    # Create an empty DataFrame
    df = pd.DataFrame()
    
    # Iterate through every 2 lines and step by 2 lines
    for i in range(1, len(lines), 2):
        # Extract the current line and the next line (if it exists)
        line1 = lines[i].strip()
        line2 = lines[i + 1].strip() if i + 1 < len(lines) else None  # Handle the case where there's an odd number of lines

        # Skip lines with specific keywords
        if "VARIABLE" in line1 or "age_cat" in line1 or "Const" in line1 or "p<0.01" in line1 or "Obser" in line2:
            continue

        # Create a new row for the DataFrame
        new_row = {"Type": line1.split()[0], 'avg ' + num: line1.split()[1], 'sd ' + num: line2.split()}

        # Append the new row to the DataFrame
        df = df.append(new_row, ignore_index=True)

        # Your code to process the lines goes here
        # For example, you can print the lines:
        print("Line 1:", line1)
        print("Line 2:", line2)
        print("------")
    
    # Return the resulting DataFrame
    return df
```
