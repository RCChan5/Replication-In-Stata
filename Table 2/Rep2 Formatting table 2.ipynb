# Formatting Stata CSV Output 
# NOTE:All the calculations are done within stata and this code reformats the output without hardcoding any variables

```python
import pandas as pd
```

## Importing the stata output as seen below

```python
# Replace 'your_file.csv' with the actual path to your CSV file
file_path = 'replication2.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)
print("This is the base dataframe that stata outputed")
df
```

## Cleaning the ="***"  from the dataframe

```python
df_c = df.applymap(lambda x: str(x).replace('=',' ').replace('"',' '))
df_c
```

## Removing unused features 

```python
last = df_c.iloc[-2]
df_c = df_c.drop([df_c.index[-1], df_c.index[0], df_c.index[-2]])
df_c
```

## Shifting F value to the top of the dataframe

```python
x=last.to_frame().T
x
```

```python
new = pd.concat([df_c.iloc[:0], x, df_c.iloc[0:]], ignore_index=True)
new
```

## Cleaning the column names

```python
new.columns = [col.replace('=', '').replace('"', '').replace('(', '').replace(')', '') for col in new.columns]
new
```

```python
# Iterate through column names and modify them
for col in new.columns:
    # Skip columns with empty names or "1"
    if col == "" or col == "1":
        continue
    
    # Modify the column name (in this example, add "_new" to each column name)
    new_col_name = int(col) + 1
    
    # Rename the column
    new.rename(columns={col: new_col_name}, inplace=True)

df = new
df
```

## Rounding to the values within the dataframe

```python
def update_value(value):
    if "." in value:
        if "(" in value:
            # Removing parentheses
            number_without_parentheses = value.replace("(", "").replace(")", "").strip()
            value = round(float(number_without_parentheses), 3)
            # Adding parentheses back
            value = f"({value})"
        else:
            value = round(float(value), 3)
    return value

# Apply the update_value function to each cell in the DataFrame
df = df.applymap(update_value)
df.iloc[0, 1:] = df.iloc[0, 1:].astype(float).round(2)
df
```
